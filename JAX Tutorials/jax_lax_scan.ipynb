```python
import jax
import jax.numpy as jnp
```

```python
def summation(initSum, xarr):
    sum = initSum
    for ele in xarr:
        sum = sum + ele
    return sum
```

```python
key = jax.random.key(0)
xarr = jax.random.randint(key, 10, 1, 45)

print(f"xarr = {xarr}")
print(f"Total Sum = {summation(0, xarr)}")
```

```python
def scan_sum(carry, xi):
    sum = carry + xi
    return sum, sum

tsum, res = jax.lax.scan(scan_sum, 0, xarr)
print(f"inp arr : {xarr}")
print(f"Total Sum = {tsum}")
print(f"Res = {res}")
```
